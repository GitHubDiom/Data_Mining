```python
import os,pandas as pd
file_name = './ml-100k/u.data'
all_ratings = pd.read_csv(file_name , delimiter='\t',names = ['UserID' , 'MovieID', 'Rating', 'Datetime'],header=None)
all_ratings['Datetime'] = pd.to_datetime(all_ratings['Datetime'],unit='s')
all_ratings[:6]
```

```python
all_ratings['favorable'] = all_ratings['Rating']>3
Train_favorable_ratings = all_ratings[all_ratings['UserID'].isin(range(200))&all_ratings['favorable']]
Test_favorable_ratings = all_ratings[all_ratings['UserID'].isin(range(200,300))&all_ratings['favorable']]
Test_favorable_ratings.head()
```

```python
'''
    1.统计每个用户喜欢过的电影
    2.统计每部电影的被喜欢数
'''
Train_favorated_by_users = {user : frozenset(movieid) for user , movieid in Train_favorable_ratings.groupby('UserID')['MovieID']}
#movies_favorable_num = favorable_ratings.groupby('MovieID').favorable.sum()
#%matplotlib inline
Train_movies_favorable_num = Train_favorable_ratings[['MovieID', 'favorable']].groupby('MovieID').count()
Train_movies_favorable_num.sort_values('favorable',ascending=False)[:5]
#movies_favorable_num.sort_values('favorable',ascending=False)[:5].plot(kind = 'Bar')
```

```python
Test_favorated_by_users = {user : frozenset(movieid) for user , movieid in Test_favorable_ratings.groupby("UserID").MovieID}
Test_movies_favorable_num = Test_favorable_ratings[['MovieID' ,'favorable']].groupby('MovieID').count()
Test_movies_favorable_num.sort_values('favorable' , ascending = False)[:5]
```

```python
Train_frequent_itemsets = {}
Test_frequent_itemsets = {}
min_support = 50
Train_frequent_itemsets[1] = dict((frozenset((movie_id,)) , row['favorable'])
                           for movie_id , row in Train_movies_favorable_num.iterrows() 
                           if row['favorable'] > min_support)
Test_frequent_itemsets[1] = { frozenset((movie_id,)) : row['favorable'] for movie_id , row in Test_movies_favorable_num.iterrows()
                            if row['favorable'] > min_support}
```

```python
from collections import defaultdict
def find_frequent_itemsets(favorated_by_users , k_1_itemsets , min_support):
    counts = defaultdict(int)
    for user , reviews in favorated_by_users.items():
        for itemset in k_1_itemsets:
            if itemset.issubset(reviews):
                for other_reviewed_movie in reviews-itemset:
                    current_superset = itemset | frozenset((other_reviewed_movie,))#若other为空集，则用当前子集代替
                    counts[current_superset]+=1
    return {itemset : frequecy for itemset , frequecy in counts.items() if frequecy >= min_support}
#find_frequent_itemsets(favorated_by_users , frequent_itemsets[1] , 50)
```

```python
for k in range(2, 20):
    Train_current_frequents_itemsets = find_frequent_itemsets(Train_favorated_by_users ,Train_frequent_itemsets[k-1] , 50)
    Train_frequent_itemsets[k] = Train_current_frequents_itemsets
    if len(Train_current_frequents_itemsets)==0:
        break
del Train_frequent_itemsets[1]
'''频繁项集是一组达到最小支持度的项目集合， 而关联规则由前提和结论组成'''
#frequent_itemsets#是频繁项集但不是真正意义上的关联规则
len(Train_frequent_itemsets)
```

```python
for k in range(2, 20):
    Test_current_frequents_itemsets = find_frequent_itemsets(Test_favorated_by_users ,Test_frequent_itemsets[k-1] , 50)
    Test_frequent_itemsets[k] = Test_current_frequents_itemsets
    if len(Test_current_frequents_itemsets)==0:
        break
del Test_frequent_itemsets[1]
```

```python
#Test_frequent_itemsets
```

```python
Train_candidate_rules = []
for itemset_len , itemset_count in Train_frequent_itemsets.items():
    for itemset in itemset_count:
        for conclusion in itemset:
            premise = itemset - set((conclusion,))
            Train_candidate_rules.append((premise , conclusion))
Train_candidate_rules[:5]
```

```python
'''创建随机森林估计器'''
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state = 14)
```

```python
import numpy as np
rule_len = 8
'''进行交叉检验时 ， 特征集的长度需要相等？
    如果遇到长度不一时，应该怎么办
'''
Train_premise_set = np.array([[int(premise) for premise in rule[0]]  for rule in Train_candidate_rules if len(rule[0]) ==rule_len ])
Train_conclusion = np.array([rule[1] for rule in Train_candidate_rules if len(rule[0]) == rule_len ] )
```

```python
'''利用交叉检验验证推荐电影模型的准确性
    用什么来检验交叉检验的正确率是否正确 - -
'''
from sklearn.cross_validation import cross_val_score
#Test_premise , Test_conclusion = Test_candidate_rules
scores = cross_val_score(clf , Train_premise_set , Train_conclusion , scoring = 'accuracy')
np.mean(scores)*100
```

```python
Train_correct_counts = defaultdict(int)
Train_incorrect_counts = defaultdict(int)
for user, favorated_movie_id in Train_favorated_by_users.items():
    for rule in Train_candidate_rules:
        premise , conclusion = rule
        if premise.issubset(favorated_movie_id):
            if conclusion in favorated_movie_id:
                Train_correct_counts[(premise , conclusion)]+=1
            else:
                Train_incorrect_counts[(premise , conclusion)]+=1
```

```python
from operator import itemgetter
'''训练数据是ID为200以内的用户点赞信息
    统计各种“规则”的置信度后进行排序
'''
Train_rule_confidence = { rule : Train_correct_counts[rule]/float(Train_incorrect_counts[rule]+Train_correct_counts[rule])
                   for rule in Train_candidate_rules}
Train_sorted_confidence = sorted(Train_rule_confidence.items() , key = itemgetter(1), reverse= True )
#rule_confidence
```
